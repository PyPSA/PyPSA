# SMS++ Optimization Backend

This example solves a PyPSA network with the optional SMS++ backend. It requires the Python packages `pypsa2smspp` and `pysmspp`, as well as the native SMS++ binaries, for example from `conda install -c conda-forge smspp-project`.

## One-Step Optimization

Use `solver_name="smspp"` to route `n.optimize()` through the SMS++ conversion, solve, and solution retrieval pipeline.

```python
import pypsa

n = pypsa.examples.ac_dc_meshed()

# Global constraints are not yet supported by the SMS++ conversion.
n.remove("GlobalConstraint", n.global_constraints.index)

status, condition = n.optimize(solver_name="smspp")
```

```python
status, condition
```

## Passing SMS++ Options

Keyword arguments for `pypsa2smspp.Transformation` can be passed through `solver_options`.

```python
status, condition = n.optimize(
    solver_name="smspp",
    solver_options={
        "workdir": "output",
        "name": "test_case",
    },
)
```
